```python
data = pd.read_csv(r"/content/Frailty Analysis Data.csv")

print(data)
```

```python
data.info()
```

```python
import matplotlib.pyplot as plot
import numpy as np
import seaborn as sns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Grip Strength', data=data, color='green', marker='o')

# Add labels and title
plt.title('Age Vs Grip Strength')
plt.xlabel('Age')
plt.ylabel('Grip Strength')

# Save the plot as a PNG file
plt.savefig(r"/content/sample_data/agevsgrip.png")

# Display the plot
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a distplot for 'Grip Strength'
sns.displot(data=data, x='Grip Strength', kde=True, color='blue', height=5, aspect=1.5)

# Add title and labels
plt.title('Distribution of Grip Strength')
plt.xlabel('Grip Strength')
plt.ylabel('Density')

# Save the plot as a PNG file
plt.savefig(r"/content/sample_data/distofgrip.png")

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(x='Height', y='Grip Strength', data=data, color='lightblue')

# Add title and labels
plt.title('Height Vs Grip Strength')
plt.xlabel('Height')
plt.ylabel('Grip Strength')

# Save the plot as a PNG file
plt.savefig(r"/content/sample_data/heightvsgrip.png")

# Display the plot
plt.show()
```

```python
sns.distplot(data['Age'], kde = False, color ='pink', bins = 30)
fig = plot.gcf()
fig.savefig(r"/content/sample_data\Age_Distribution.png")
```
